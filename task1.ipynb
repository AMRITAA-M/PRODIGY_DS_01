```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("genderdistr.csv")
# Display first few rows
print(df.head())
```

```python
# Plot Bar Chart for Gender Distribution
df.plot(x="Year", y=["Male Population", "Female Population"], kind="bar", figsize=(12,6))
plt.title("Gender Distribution in a Population (CATEGORICAL VARIABLE)", fontsize=20, fontweight="bold")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend(["Male", "Female"])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("agedistr.csv")
print(df.head())

# Set Seaborn style
sns.set_style("darkgrid")

# Create figure
plt.figure(figsize=(12,6))

# Plot Histogram
sns.barplot(x="Age Group", y="Population", data=df, palette="plasma", edgecolor="black")

# Labels and Title
plt.xlabel("Age Group", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.title("Age Distribution in a Population (CONTINUOUS VARIABLE)", fontsize=20, fontweight="bold")

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
